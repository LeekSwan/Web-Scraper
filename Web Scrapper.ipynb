```python
def scroll(driver, timeout): #func to scrape infinite scrolling pages. 

    last_height = driver.execute_script("return document.body.scrollHeight") #finds scroll height
    
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") #scroll to bottom
        time.sleep(timeout) #load page
        time
        #calcualte new scroll height compared to old height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            # If heights are the same it will exit the function
            break
        last_height = new_height
        
```

```python
def get_html(url):
    
    driver.get(url)
    scroll(driver, 5)
    soup = bs(driver.page_source, "lxml")
    driver.close
    #print(soup)
    
    return soup
soup = get_html(req_url)
```
